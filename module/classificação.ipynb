Carregando o módulos e os dados a serem utilizados

```python
from pandas import read_csv, DataFrame, Series, get_dummies
from platform import system
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from geopy import geocoders
from numpy import where, NaN, array

data = read_csv('file/train.csv', encoding='latin1');
dp = DataFrame(index=data.index.tolist(), columns=data.columns[ (data.columns != 'city') &(data.columns != 'region') &(data.columns != 'item_id') &(data.columns != 'user_id') &(data.columns != 'image') &(data.columns != 'image_top_1')].tolist());
dp['title'] = data['title'];
dp['description'] = data['description'];
dp['price'] = data['price'];
dp['item_seq_number'] = data['item_seq_number'];
dp['activation_date'] = data['activation_date'];
dp['deal_probability'] = data['deal_probability'];
```

"Cruzando" as cidades e as regiões russas para obter coordenadas e adicionar ao DataFrame(dp), onde possui os dados já processados.

```python
cityRegion = data.loc[:, ["city", "region"]].apply(lambda l: " ".join(l), axis=1)
g = geocoders.Nominatim();
coord = [];
for location in cityRegion:
    geocode = g.geocode("Самара Самарская область", timeout=10, language='pt');
    coord.append([geocode.latitude, geocode.longitude]);
    break;
    #if location != NaN:
        #geocode = g.geocode(location, timeout=10, language='en');
        #coord.append([geocode.latitude, geocode.longitude]);
    #else: coord.append(NaN);
#dp['city_region'] = coord;
print(coord);
```

Obtendo as classificações, através do hot encoded

```python
heads = ['parent_category_name', 'category_name', 'param_1', 'param_2', 'param_3', 'user_type'];
for head in heads:  dp[head] = list(get_dummies(data[head]).values);
print(dp);
```
